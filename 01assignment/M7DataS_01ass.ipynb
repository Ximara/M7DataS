```python
numeric = df.select_dtypes(include=[np.number])
numeric.columns
```

```python
yearbuilt = np.array(numeric['YearBuilt'])
saleprice = np.array(numeric['SalePrice'])
yearsold = np.array(numeric['YrSold'])
lotarea = np.array(numeric['LotArea'])
```

```python
plt.hist(yearbuilt)
plt.xticks(np.arange(1870, 2020, step=10), rotation='vertical')
plt.title('Number of houses built in decade')
plt.show()
```

```python
plt.scatter(yearbuilt, saleprice)
plt.xlabel('Year Built')
plt.ylabel('Sale Price')
plt.title('Sale Price of houses built in year')
plt.show()
```
